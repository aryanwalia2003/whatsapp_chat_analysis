```python

# Define the regex pattern for the date and time format, including the optional non-breaking space
pattern = r"\b\d{2}/\d{2}/\d{2}, \d{1,2}:\d{2}[\u202F\s]?[ap]m\b"

# Use regex to find all matches of the date and time format
dates = re.findall(pattern, data)

# Split the text based on the matches
messages = re.split(pattern, data)

# Remove empty elements from the messages list
messages = [message.strip() for message in messages if message.strip()]

```

```python

from datetime import datetime
# Function to convert date string to datetime object
def convert_to_datetime(date_str):
    return datetime.strptime(date_str, "%d/%m/%y, %I:%M %p")

# Convert and add all the dates back to the list
for i, date_str in enumerate(dates):
    dt_obj = convert_to_datetime(date_str)
    dates[i] = dt_obj.strftime('%d-%m-%y %H:%M:%S')
```

```python
def separate_messages_and_users(messages):
    users = []
    messages_list = []

    for message in messages:
        # Check if the colon is present in the message
        if ':' in message:
            # Split the message by the first colon encountered
            colon_index = message.index(':')
            user = message[:colon_index].strip()
            msg = message[colon_index + 1:].strip()

            users.append(user)
            messages_list.append(msg)
        else:
            # If the colon is not found, treat the entire message as a user and an empty message
            users.append(message)
            messages_list.append('')

    return users, messages_list
```

```python
users_list, messages_list = separate_messages_and_users(messages)
```

```python
import pandas as pd

# Assuming you have a pandas DataFrame called df with a "User" column
# You can use regular expressions to define patterns for unwanted messages

# Define a list of patterns to match admin messages, join messages, and "added" messages
unwanted_patterns = [
    r"added \+\d+",
    r"joined using this",
    r"added \w+",
    r"removed \+\d+",
    r"removed \w+",
    r"changed",
    r"updated",
    r"turned",
    r"created",
    r"Messages and calls",
    r"deleted"
]

# Combine the patterns using the OR (|) operator
pattern = '|'.join(unwanted_patterns)

# Use the str.contains() method to create a boolean mask
mask = df['User'].str.contains(pattern, case=False)

# Apply the mask to filter out the unwanted rows and get only the valid users
filtered_users = df.loc[~mask, 'User'].unique()

# filtered_users will now contain an array of unique user names without the unwanted messages
print(filtered_users)
print(len(filtered_users))
```

```python
import nltk
import string

def pre_processing(text):
    text = text.lower()  # Convert into lower case
    words = nltk.word_tokenize(text)  # Break into words
    clean_words = []

    for word in words:
        if word.isalnum():  # Removing special characters
            clean_words.append(word)

    # Removing stop words and punctuations
    clean_words = [word for word in clean_words if word not in string.punctuation]

    return " ".join(clean_words)

# Test the function
input_text = "- +91 93545 89727: A notice is being circulated on many WhatsApp group which states that Itâ€™s off on Monday because of international poverty day. I want to inform you all that the notice is FAKE and itâ€™s a working day on Monday."
processed_text = pre_processing(input_text)
print(processed_text)
```
